## Conversion of the age column

If we have a look at the age column we can see that the values are not in years, instead they are in days. Now since the numbers are quite large and high, the model might have some problem in understanding that column properly. So the better option will be that we convert the days into years. 

To do the above we will need this formula:

                                         Age (years)= Number of days/365.25
 



```python
df['Age'] = df['Age']/365.25
df['Age'] = df['Age'].round(0) 
```

```python
from imblearn.over_sampling import RandomOverSampler

X = df.drop('Status', axis=1) 
y = df['Status']  

oversampler = RandomOverSampler(random_state=42)

X_resampled, y_resampled = oversampler.fit_resample(X, y)

df_resampled = pd.concat([X_resampled, y_resampled], axis=1)

print(df_resampled['Status'].value_counts())
```

```python
df_resampled.shape
```

```python
df.shape
```

#### Note: In the realistic world, oversampling is supposed to be applied on the training set only to avoid data leak

```python
df_resampled.head()
```

```python
df_resampled.info()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingClassifier
```

```python
X = df_resampled.drop('Status',axis=1)
y = df_resampled['Status']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
from sklearn.metrics import accuracy_score, classification_report
```

```python
model = GradientBoostingClassifier()

# Train the model on the training data
model.fit(X_train, y_train)

# Make predictions on the test data
y_pred = model.predict(X_test)

# Evaluate the model's performance on the test data
accuracy = accuracy_score(y_test, y_pred)
classification_report_result = classification_report(y_test, y_pred)

print("Accuracy:", accuracy)
print("Classification Report:\n", classification_report_result)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from xgboost import XGBClassifier
```

```python
# RandomForestClassifier
rf_model = RandomForestClassifier()
rf_model.fit(X_train, y_train)
rf_pred = rf_model.predict(X_test)
rf_accuracy = accuracy_score(y_test, rf_pred)
rf_classification_report = classification_report(y_test, rf_pred)

# DecisionTreeClassifier
dt_model = DecisionTreeClassifier()
dt_model.fit(X_train, y_train)
dt_pred = dt_model.predict(X_test)
dt_accuracy = accuracy_score(y_test, dt_pred)
dt_classification_report = classification_report(y_test, dt_pred)



# Print results for each model
print("RandomForestClassifier Accuracy:", rf_accuracy)
print("RandomForestClassifier Classification Report:\n", rf_classification_report)

print("\nDecisionTreeClassifier Accuracy:", dt_accuracy)
print("DecisionTreeClassifier Classification Report:\n", dt_classification_report)
```
